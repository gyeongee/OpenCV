```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 허프선 변환
img = cv2.imread('./img/sudoku.jpg')
img2 = img.copy()
h,w = img.shape[:2]

imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(imgray,100,100)


# lines = cv2.HoughLines(image, rho, theta, threshold)

lines = cv2.HoughLines(edges,1,np.pi/180,130)
for line in lines:
    r,theta = line[0] # 거리와 각도 w,h
    tx,ty = np.cos(theta),np.sin(theta)
    x0,y0 = tx*r, ty*r
    # 기존 좌표에 빨간원
    # cv2.circle(image, center, radius, color, thickness)

    cv2.circle(img2, (int(x0), int(abs(y0))), int(abs(y0)), (0, 0, 255), -1)

    #  직선의 방정식으로 그리기 위한 두 개의 점(시작점과 끝점)
    # w = 너비
    x1,y1 = int(x0 + w * (-ty)), int(y0 + h * tx)
    x2,y2 = int(x0 - w * (-ty)), int(y0 - h * tx)
    # 선 그리기
    cv2.line(img2, (x1,y1), (x2,y2), (0,255,0),1)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(10,6))
plt.subplot(121), plt.axis('off'), plt.imshow(img), plt.title('original')
plt.subplot(122), plt.axis('off'), plt.imshow(img2), plt.title('hough line')
plt.show()
```

```python
# 확률적 허프 선 변환
img = cv2.imread('./img/sudoku.jpg')
img2 = img.copy()
h,w = img.shape[:2]

imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(imgray,50,200)

lines = cv2.HoughLinesP(edges,1,np.pi/180,10,None,20,2)

for line in lines:
    x1,y1,x2,y2 = line[0]
    cv2.line(img2,(x1,y1),(x2,y2),(0,255,0),1)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(10,6))
plt.subplot(121), plt.axis('off'), plt.imshow(img), plt.title('original')
plt.subplot(122), plt.axis('off'), plt.imshow(img2), plt.title('hough lineP')
plt.show()
```

```python
# 허프 원 변환
img = cv2.imread('./img/circles.png')
origin = img.copy()
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray,(3,3),0)

# default 값으로 파라미터 조정
# circles = cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,1.5,30,None,200)

circles = cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,1.2,40)

if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        # 원둘레
        cv2.circle(img, (i[0], i[1]), i[2], (0, 255, 0), 2)
        # 원 중심
        cv2.circle(img,(i[0],i[1]),2,(0,0,255),5)

origin = cv2.cvtColor(origin, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(10,6))
plt.subplot(121), plt.axis('off'), plt.imshow(origin), plt.title('original')
plt.subplot(122), plt.axis('off'), plt.imshow(img), plt.title('hough Circle')
plt.show()
        
```

```python
# 거리 변환
# 연속된 영역을 찾아 분할하는 방법
img = cv2.imread('./img/arrow.jpg', cv2.IMREAD_GRAYSCALE)
print(img.shape)
_,biimg = cv2.threshold(img, 127,255,cv2.THRESH_BINARY_INV)

dist = cv2.distanceTransform(biimg,cv2.DIST_L2,5)
# 정규화 0~255
dist = (dist/(dist.max()-dist.min())*255).astype(np.uint8)

skel = cv2.adaptiveThreshold(dist,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                            cv2.THRESH_BINARY,7,-3)
print(skel.shape)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
biimg = cv2.cvtColor(biimg, cv2.COLOR_BGR2RGB)
dist = cv2.cvtColor(dist, cv2.COLOR_BGR2RGB)
skel = cv2.cvtColor(skel, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(10,6))
plt.subplot(221), plt.axis('off'), plt.imshow(img), plt.title('original')
plt.subplot(222), plt.axis('off'), plt.imshow(biimg), plt.title('biimg')
plt.subplot(223), plt.axis('off'), plt.imshow(dist), plt.title('distance')
plt.subplot(224), plt.axis('off'), plt.imshow(skel), plt.title('skeleton')
plt.show()
        
```

```python
# 색 채우기
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('./img/taekwonv1.jpg')
img2 = img.copy()
rows,cols = img.shape[:2]

mask = np.zeros((rows+2, cols+2),np.uint8)

newVal = (255,255,255)

loDiff,upDiff = (10,10,10),(10,10,10)

def onMouse(event,x,y,flags, param):
    global mask, img2
    if event == cv2.EVENT_LBUTTONDOWN:
        seed = (x,y)
        retval = cv2.floodFill(img2,mask,seed,newVal,loDiff,upDiff)
        cv2.imshow('img',img2)

cv2.imshow('img',img2)
cv2.setMouseCallback('img',onMouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread('./img/taekwonv1.jpg')
img_draw = img.copy()
marker = np.zeros(img.shape[:2], dtype=np.int32)  # 'masker' → 'marker'

markerId = 1
colors = []
isDragging = False

def onMouse(event, x, y, flags, param):
    global img_draw, marker, markerId, isDragging, colors

    if event == cv2.EVENT_LBUTTONDOWN:
        isDragging = True
        # 현재 마커 아이디에 해당하는 색상 저장 (중복 방지)
        if markerId > len(colors):
            colors.append(tuple(int(c) for c in img[y, x]))

    elif event == cv2.EVENT_MOUSEMOVE:
        if isDragging:
            marker[y, x] = markerId
            cv2.circle(img_draw, (x, y), 3, (0, 0, 255), -1)
            cv2.imshow('watershed', img_draw)

    elif event == cv2.EVENT_LBUTTONUP:
        if isDragging:
            isDragging = False
            markerId += 1

    elif event == cv2.EVENT_RBUTTONDOWN:
        # watershed 적용
        cv2.watershed(img, marker)
        img_result = img.copy()

        # 경계선 (-1): 초록색
        img_result[marker == -1] = (0, 255, 0)

        # 각 마커 색상 적용
        for idx, color in enumerate(colors):
            img_result[marker == (idx + 1)] = color

        cv2.imshow('watershed', img_result)

cv2.imshow('watershed', img)
cv2.setMouseCallback('watershed', onMouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# 이미지 읽기
img = cv2.imread('./img/taekwonv1.jpg')
img_draw = img.copy()

marker = np.zeros(img.shape[:2], dtype=np.int32)

# 마커 ID, 마커 색상 저장 리스트
markerId = 1
colors = []

# 마우스 드래그 상태
isDragging = False

def onMouse(event, x, y, flags, param):
    global img_draw, marker, markerId, isDragging, colors

    if event == cv2.EVENT_LBUTTONDOWN:
        isDragging = True
        # 현재 위치의 픽셀 색상을 저장 (마커 ID와 함께)
        colors.append((markerId, img[y, x].tolist()))

    elif event == cv2.EVENT_MOUSEMOVE:
        if isDragging:
            marker[y, x] = markerId
            cv2.circle(img_draw, (x, y), 3, (0, 0, 255), -1)
            cv2.imshow('watershed', img_draw)

    elif event == cv2.EVENT_LBUTTONUP:
        if isDragging:
            isDragging = False
            markerId += 1

    elif event == cv2.EVENT_RBUTTONDOWN:
        # 워터셰드 알고리즘 적용
        cv2.watershed(img, marker)
        img_draw[marker == -1] = (0, 255, 0)  # 경계선은 초록색으로 표시

        # 각 마커 ID에 해당하는 색상으로 색칠
        for mid, color in colors:
            img_draw[marker == mid] = color

        cv2.imshow('watershed', img_draw)

# 창 생성 및 마우스 콜백 설정
cv2.imshow('watershed', img_draw)
cv2.setMouseCallback('watershed', onMouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 그랩컷
# 동일 레이블을 갖는 연결된 영역의 객체 분리
img = cv2.imread('./img/taekwonv1.jpg')
img_draw = img.copy()
marker = np.zeros(img.shape[:2], dtype=np.int32)
rect=[0,0,0,0]
mode = cv2.GC_EVAL
bgdmodel = np.zeros((1,65),np.float64)
fgdmodel = np.zeros((1,65),np.float64)

def onMouse(event, x, y, flags, param):
    global rect, marker,mode
    if event == cv2.EVENT_LBUTTONDOWN:
        if flags <= 1:
            mode = cv2.GC_INIT_WITH_RECT
            rect[:2] = x,y
    elif event == cv2.EVENT_MOUSEMOVE and flags&cv2.EVENT_FLAG_LBUTTON:
        if mode == cv2.GC_INIT_WITH_RECT:
            img_temp = img.copy()
            cv2.rectangle(img_temp,(rect[0],rect[1]),(x,y),(0,255,0),2)
            cv2.imshow('img',img_temp)
        elif flags > 1:
            mode = cv2.GC_INIT_WITH_MASK:
            if flags & cv2.EVENT_FLAG_CTRLKEY:
                cv2.circle(img_draw,(x,y),3,(255,255,255),-1)
                cv2.circle(marker,(x,y),3,cv2.GC_FGD,-1)
            if flags & cv2.EVENT_FLAG_SHIFTEY:
                cv2.circle(img_draw,(x,y),3,(0,0,0),-1)
                cv2.circle(marker,(x,y),3,cv2.GC_BGD,-1)
            cv.imshow('img',img_draw)
        elif event == cv2.EVENT_LBUTTONUP:
            if mode == cv2.GC_INIT_WITH_RECT:
                rect[2:] = x,y
                cv.rectangle(img_draw,(rect[0],rect[1],),(rect[2],rect[3]),(255,0,0))
                cv2.imshow('img',img_draw)
            # 그랩컷
            cv2.grabCut(img,mask,tuple(rect),bgdmodel,fgdmodel,1,mode)
            img2 = img.coppy()
            # 확실한 배경 혹은 아마도 배경이면 표시된 영역을 0으로 채우기
            img[(mask==cv2.GC_BGD) | (mask==cv2.GC_PR_BGD)]=0
            cv2.imshow('grabcut',img2)
            mode = cv2.GC_EVAL
cv2.imshow('img',img)
cv2.setMouseCallback('img',onMouse)
while True:
    if cv2.waitKey(0) & 0xFF == 27:
        break
    else:
        
            
                
```
